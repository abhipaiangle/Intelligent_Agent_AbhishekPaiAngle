1. Romance

7. Breakup

##Build of our app
We shall consier every song as a 7 dimensional vector, and has values for each vector. The sum of coefficients of each vector is 1. Every song also has it's 
Now when the person starts the app: 
**Initialize a list of played songs nd a list of blacklisted(Skipped) songs**
1. He types a song and the song is played.
2. We enter the song in our input
3. Break the song into it's genre's and get coefficients
4. Let the coefficients be g1, g2, g3, g4, g5, g6, g7
5. Calculate probabilities of each song using these coefficients.(Simple regression or ANN)
6. Output the next song
7. If the person skips the song add the song to blacklist. 
8. if it's okay add the song to played.
9. Blacklisted songs are not to be played for the user

**Selecting the song to be suggested based on the previous song** 

We shall use the history of the first 5 songs: 
1. The first song shall be selected by the user by typing the name. It shall be inserted in the played list
2. For recommending the second song use the first one 
3. For third song use the first 2 songs and take the average of probability
4. For fourth use first 3
5. For fifth use first 4
6. For sixth use first 5
7. From here use the last 5
8. If a song is skipped do not use it in suggesting next
